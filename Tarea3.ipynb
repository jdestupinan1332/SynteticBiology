#### Programe en Matlab (u otro lenguaje) una simulación de la expresión de un gen con tasa de transcripción constante, tasa de traducción constante por mRNA, dilución por proteína y destrucción por mRNA constantes (correspondientes a $\dot{r} = k_r − \gamma_r r $ y $\dot{p} = k_p r − \gamma_p p$ ), usando la aproximación de intervalos de tiempo $\Delta t$  pequeños con tasas constantes de creación y destrucción por intervalo. Escoja constantes biológicamente relevantes.

```python
def simRNA (k_r, gamm_r, delt, tsim, r0):
    #Definicion de constantes
    kr = k_r
    gammr = gamm_r
    dt = delt
    sim_t = tsim
    t=0.0
    
    r = 0
    listr = [r0]   

    while(t<= sim_t):
        rand = random.random(2) #lanza numeros aleatorios
        if(rand[0] < kr*dt): #condicion para que cree
            r = r + 1
        if(rand[1] < gammr*r*dt): #condicion para que destruya
            r = r - 1
        listr.append(r)
        t=t+dt
        
    return listr
```

```python
def simPRO (k_p, gamm_p, delt, tsim, listr, p0):
    #definicion de constantes
    kp = k_p
    gammp = gamm_p
    dt = delt
    sim_t = tsim
    t=0.0
    
    p = 0
    i = 0
    listp = [p0]
    
    while (t<=sim_t):
        rand = random.random(2) #lanza numeros aleatorios
        if(rand[0] < kp*listr[i]*dt): #condicion para que cree
            p = p + 1
        if(rand[1] < gammp*p*dt): #condicion para que destruya
            p = p - 1
        listp.append(p)
        t=t+dt
        i=i+1
        
    return listp
```

```python
rna = simRNA(1.0/1, 1.0/5, 1.0/1000, 1000.0, 0)
pro = simPRO(33.0/1, 1.0/30, 1.0/1000, 1000.0, rna, 0)
```

#### Cambie su simulación para incluir retroalimentación negativa (correspondiente a $\dot{r} = \frac{k_r}{1+ P^2/ K} - \gamma_r r$ y $\dot{p}=k_p r- \gamma_p p$) y repita el punto b)

```python
def simFeed (k_r, k_p, K, gamm_r, gamm_p, delt, tsim, r0, p0):
    #Definicion de constantes
    kr = k_r
    gammr = gamm_r
    kp = k_p
    gammp = gamm_p
    
    dt = delt
    sim_t = tsim
    t=0.0

    i = 0
    r = r0
    p = p0
    listr = [r]
    listp = [p]

    while(t<= sim_t):
        rand = random.random(4) #lanza numeros aleatorios
        if(rand[0] < kr*dt/(1 + (listp[i]^2*K))): #condicion para que cree
            r = r + 1
        if(rand[1] < gammr*r*dt): #condicion para que destruya
            r = r - 1
        listr.append(r)
        if(rand[2] < kp*listr[i]*dt): #condicion para que cree
            p = p + 1
        if(rand[3] < gammp*p*dt): #condicion para que destruya
            p = p - 1
        listp.append(p)
        i = i+1
        t=t+dt
            
    return listr, listp
```

# Simulación De Gillespie

#### Programe en Matlab (u otro lenguaje) una simulación de la expresión de un gen con tasa de transcripción constante, tasa de traducción constante por mRNA, dilución por proteína y destrucción por mRNA constantes (correspondientes a $\dot{r} = k_r − \gamma_r r $ y $\dot{p} = k_p r − \gamma_p p$), usando el algoritmo de Gillespie. Escoja constantes biológicamente relevantes (las mismas d la tarea anterior).

```python
def gill(k_r, k_p, gamm_r, gamm_p, tsim, r0, p0):
    #inicializacion de constantes
    kR = k_r
    kP = k_p
    gammR = gamm_r
    gammP = gamm_p
    t = 0.0
    
    
    #listas
    listR = [r0]
    listP = [p0]
    listT = [0]
    
    t = 0
    r = r0
    p = p0
    i = 0
    
    while(t<=tsim):
        K = kR + gammR*listR[i] + kP*listR[i] + gammP*listP[i]
        u = random.random(2) #numeros aleatorios
        talgo = 1/K *np.log(1/u[0]) #tiempo para que suceda algo
        #print u[1]
        if (u[1] <  kR/K): #crea RNA
            #print ('CR')
            r = r+1
            p = p
        elif (u[1] < (kR+gammR*listR[i])/K): #destruye RNA
            #print ('DR')
            r = r-1
            p = p
        elif (u[1] < (kR+gammR*listR[i]+kP*listR[i])/K): #crea Prot por burst
            #print ('CP')
            r = r
            p = p+1
        else: #destruye Prot por burst
            #print ('DP')
            r = r
            p = p-1
            
        listR.append(r)
        listP.append(p)
        listT.append(t+talgo)
        t=t+talgo
        i = i+1
    
    return listR, listP, listT
```

```python
rna, pro, tiempo = gill(1/1.0, 33.0/1, 1/5.0, 1/30.0, 200, 0,0)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(tiempo, rna, '-', label = 'RNA')
plt.legend(loc=2)
ax2 = ax.twinx()
ax2.plot(tiempo, pro, '-g', label = 'Protein')
ax.set_xlabel("Time")
ax.set_ylabel("RNA")
ax2.set_ylabel("Protein")
ax2.set_ylim(0, 10000)
ax.set_ylim(0,30)
plt.legend()
plt.show()
```
